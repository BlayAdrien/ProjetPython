```python
dataFinal=pd.DataFrame(data.YearsProgram)
dataFinal.insert(1,"Education",pd.DataFrame(data.FormalEducation))
dataFinal.insert(2,"Salary",pd.DataFrame(data.Salary))
#dataFinalMajor=pd.DataFrame(dataFinal)
#dataFinalMajor.insert(2,"MajorSub",pd.DataFrame(data.MajorUndergrad))
dataFinal=dataFinal.dropna()
```

```python
dataFinal=dataFinal.replace("Less than a year",0 )
dataFinal=dataFinal.replace("1 to 2 years",1 )
dataFinal=dataFinal.replace("2 to 3 years",2 )
dataFinal=dataFinal.replace("3 to 4 years",3 )
dataFinal=dataFinal.replace("4 to 5 years",4 )
dataFinal=dataFinal.replace("5 to 6 years",5 )
dataFinal=dataFinal.replace("6 to 7 years",6 )
dataFinal=dataFinal.replace("7 to 8 years",7 )
dataFinal=dataFinal.replace("8 to 9 years",8 )
dataFinal=dataFinal.replace("9 to 10 years",9 )
dataFinal=dataFinal.replace("10 to 11 years",10 )
dataFinal=dataFinal.replace("11 to 12 years",11 )
dataFinal=dataFinal.replace("12 to 13 years",12 )
dataFinal=dataFinal.replace("13 to 14 years",13 )
dataFinal=dataFinal.replace("14 to 15 years",14 )
dataFinal=dataFinal.replace("15 to 16 years",15 )
dataFinal=dataFinal.replace("16 to 17 years",16 )
dataFinal=dataFinal.replace("17 to 18 years",17 )
dataFinal=dataFinal.replace("18 to 19 years",18 )
dataFinal=dataFinal.replace("19 to 20 years",19 )
dataFinal=dataFinal.replace("20 or more years",20 )
```

```python
dataFinal=dataFinal.replace("I never completed any formal education",0 )
dataFinal=dataFinal.replace("Primary/elementary school",1 )
dataFinal=dataFinal.replace("Secondary school",2 )
dataFinal=dataFinal.replace("Some college/university study without earning a bachelor's degree",3 )
dataFinal=dataFinal.replace("Bachelor's degree",4 )
dataFinal=dataFinal.replace("Professional degree",5 )
dataFinal=dataFinal.replace("Master's degree",6 )
dataFinal=dataFinal.replace("Doctoral degree",7 )
dataFinal=dataFinal.replace("I prefer not to answer",np.nan )
```

```python
dataFinal=dataFinal.dropna()
```

```python
dataFinalX = dataFinal.to_numpy()[:, 0:2]
dataFinalSal=dataFinal.to_numpy()[:,2]
j=0
for i in dataFinalSal:
    if i>150000:
        dataFinalSal[j]=0
    elif 100000<i and i<=150000:
        dataFinalSal[j]=1
    elif 50000<i and i<=100000:
        dataFinalSal[j]=2
    else:
        dataFinalSal[j]=3
        
    j+=1
    
#dataFinal=dataFinal.drop(columns=["catSal"])
dataFinal.insert(3,"catSal",dataFinalSal)
#Les valeurs des classes
dataFinalY=dataFinal.get("catSal").values
```

```python

catSalBin=np.empty(dataFinal.catSal.shape)
j=0
for i in dataFinal.catSal:
    if i>1:
        catSalBin[j]=1
    else:
        catSalBin[j]=0
    j+=1
print(catSalBin)
dataFinal.insert(3,"catSalBin",catSalBin)
```

```python
from keras.models import Sequential
from keras.layers import Dense
#input (X) and output (y) variables
X = dataFinal.to_numpy()[:,0:2] #Les annÃ©es d'expÃ©rience et la catÃ©gorie d'Ã©tudes
y = dataFinal.to_numpy()[:,3] #La catÃ©gorie binaire de salaire, >100k et <100k
# define the keras model
# 3 couches de neurones
model = Sequential()
model.add(Dense(30, input_dim=2, activation='relu')) #30 neurones Rectified Linear Unit (ReLU) : renvoie une valeur selon 
#son poids
model.add(Dense(20, activation='relu')) #20 neurones Rectified Linear Unit (ReLU)
model.add(Dense(1, activation='sigmoid'))  #1 sortie, 0 ou 1
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) # paramÃ©trer le processus d'apprentissage
# fit the keras model on the dataset
model.fit(X, y, epochs=100, batch_size=20,verbose=1)
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))


predictions = model.predict_classes(X)
# Exemple des 50 premiers cas
for i in range(50):
	print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```
